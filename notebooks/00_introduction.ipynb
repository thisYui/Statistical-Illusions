# Statistical Illusions: When Data Misleads Rational Thinking

## 1. Motivation

Trong phân tích định lượng, các con số thường được xem như đại diện khách quan của thực tại. Giá trị trung bình, tương quan, xu hướng hay mức ý nghĩa thống kê được diễn giải như bằng chứng chắc chắn. Tuy nhiên, rất nhiều sai lầm nghiêm trọng trong kinh tế, tài chính và khoa học dữ liệu xuất phát từ việc hiểu sai các thống kê tổng hợp này.

Vấn đề không nằm ở phép tính sai, mà nằm ở việc **diễn giải sai cấu trúc**.

## 2. Central Research Question

Project này tập trung vào câu hỏi:

> Trong những điều kiện cấu trúc nào, các công cụ thống kê tiêu chuẩn tạo ra ảo giác thay vì phản ánh thực tại?

Ta không tìm cách chứng minh rằng thống kê là sai.
Ngược lại, ta muốn hiểu **khi nào nó đúng về mặt toán học nhưng sai về mặt cấu trúc**.

## 3. Scope & Non-Goals

Project này:

- Không nhằm mục tiêu dự báo.
- Không xây dựng mô hình machine learning.
- Không phân tích dữ liệu thực tế phức tạp ở giai đoạn đầu.

Thay vào đó, ta sử dụng dữ liệu tổng hợp (synthetic data) với cơ chế sinh dữ liệu được định nghĩa rõ ràng, nhằm kiểm soát hoàn toàn “ground truth”.

Việc này cho phép ta phân biệt rõ ràng giữa:
- Hiệu ứng cấu trúc
- Nhiễu thống kê
- Sai lệch do aggregation

## 4. Illusions to Be Examined

Project sẽ lần lượt phân tích:

1. Simpson’s Paradox (đảo chiều khi gộp nhóm)
2. Mean vs Median Illusion (trung bình không đại diện cho điển hình)
3. Survivorship Bias (thiên lệch do loại bỏ thất bại)
4. Spurious Correlation (tương quan do biến ẩn)
5. Instability under small samples (ảo giác từ mẫu nhỏ)

Mỗi hiện tượng sẽ được phân tích theo cấu trúc:
- Cơ chế sinh dữ liệu
- Quan sát trực quan
- Phân tích định lượng
- Giải thích cấu trúc

## 5. Epistemic Objective

Mục tiêu cuối cùng của project không phải là học thêm một danh sách bias.

Mục tiêu là xây dựng một trực giác sâu hơn:

> Thống kê là công cụ mạnh,
> nhưng nó không tự mang ý nghĩa nếu không hiểu cấu trúc sinh dữ liệu phía sau.

Khi hoàn thành project này, người đọc nên có khả năng tự hỏi:

- Giá trị trung bình này đại diện cho điều gì?
- Có biến ẩn nào đang bị che giấu không?
- Aggregation có làm thay đổi quan hệ không?
- Phân phối có lệch khiến expectation mất ý nghĩa không?

Chỉ khi đó, ta mới có thể phân biệt giữa dữ liệu phản ánh thực tại và dữ liệu tạo ra ảo giác.
