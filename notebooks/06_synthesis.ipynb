# Synthesis: Structural Limits of Statistical Inference

Trong các notebook trước, ta đã quan sát 5 dạng statistical illusion.

Mỗi illusion không phải là lỗi tính toán,
mà là hệ quả tất yếu của cấu trúc xác suất và lý thuyết ước lượng.

Mục tiêu của notebook này là:

- Phân loại các cơ chế distortion
- So sánh bản chất của chúng
- Rút ra nguyên tắc tổng quát về giới hạn của inference

## 1. Five Mechanisms of Statistical Distortion

### (1) Aggregation Distortion
Simpson’s Paradox
→ Quan hệ thay đổi khi gộp nhóm.

### (2) Distributional Distortion
Mean vs Median
→ Moment không đại diện cho typical outcome.

### (3) Selection Distortion
Survivorship Bias
→ Mẫu quan sát không đại diện population.

### (4) Structural Omission
Spurious Correlation
→ Biến ẩn tạo quan hệ giả.

### (5) Estimation Variance
Small Sample Instability
→ Estimator dao động mạnh trong mẫu nhỏ.

## 2. Structural Comparison

| Illusion | Source | Population-level? | Sample-level? |
|-----------|--------|------------------|---------------|
| Simpson | Aggregation | ✔ | ✔ |
| Mean vs Median | Skewness | ✔ | ✔ |
| Survivorship | Selection | ✔ | ✔ |
| Spurious | Omitted variable | ✔ | ✔ |
| Small sample | Estimation variance | ✘ | ✔ |

Khác biệt quan trọng:

> Small sample instability không xuất phát từ cấu trúc population, mà từ sampling distribution của estimator.

## 3. Unified View

Mọi statistical illusion đều xuất phát từ một trong ba tầng:

### Level 1 — Distribution
P(X)

### Level 2 — Structure
Conditional dependence structure

### Level 3 — Estimation
Sampling distribution of θ̂

Illusion có thể phát sinh ở bất kỳ tầng nào.

## 4. Core Insight

Statistical calculations có thể hoàn toàn chính xác.

Illusion không đến từ sai phép tính.

Nó đến từ:

- Giả định cấu trúc sai
- Quan sát mẫu không đại diện
- Hoặc giới hạn của estimator

Do đó:

Correct computation
≠
Correct interpretation
